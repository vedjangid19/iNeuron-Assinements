An instance of the class Sudoku wil have three methods:

   - get_row(n): will return the row in position n.
   - get_col(n): will return the column in position n.
   - get_sqr([n, m]): will return the square in position n if only one argument is given, and the square to which the cell in position (n, m) belongs to if two arguments are given.

```python
import numpy as np

class Sudoku:
    def __init__(self, number_string):
        self.number_string = number_string
        self.board = np.array(list(map(int, list(self.number_string)))).reshape(9,9)#.tolist()  
        
    def get_row(self, row):
        self.row = self.board[row]
        return self.row.tolist()
    
    def get_col(self, col):
        self.col = np.transpose(self.board)[col]
        return self.col.tolist()
    
    def get_sqr(self, *arg):
        
        self.squre_number = None
        
        if len(arg)>1:
            self.row = arg[0]
            self.col = arg[1]
        else:
            self.squre_number = arg[0]
        try:   
            if -1 < self.row < 3:
                if -1 < self.col < 3:
                    self.squre_number = 0
                elif 2 < self.col < 6:
                    self.squre_number = 1
                elif 5 < self.col < 9:
                    self.squre_number = 2

            elif 2 < self.row < 6:            
                if -1 < self.col < 3:
                    self.squre_number = 3
                elif 2 < self.col < 6:
                    self.squre_number = 4
                elif 5 < self.col < 9:
                    self.squre_number = 5

            elif 5 < self.row < 9:
                if -1 < self.col < 3:
                    self.squre_number = 6
                elif 2 < self.col < 6:
                    self.squre_number = 7
                elif 5 < self.col < 9:
                    self.squre_number = 8
        except  Exception as e:
            pass

        
        if self.squre_number == 0:
            self.row_lower_range = 0
            self.row_higher_range = 3
            self.col_lower_range = 0
            self.col_higher_range = 3
                
            return self.board[self.row_lower_range:self.row_higher_range, self.col_lower_range:self.col_higher_range]
            
        elif self.squre_number == 1:
            self.row_lower_range = 0
            self.row_higher_range = 3
            self.col_lower_range = 3
            self.col_higher_range = 6
                
            return self.board[self.row_lower_range:self.row_higher_range, self.col_lower_range:self.col_higher_range]
            
        elif self.squre_number == 2:
            self.row_lower_range = 0
            self.row_higher_range = 3
            self.col_lower_range = 6
            self.col_higher_range = 9
                
            return self.board[self.row_lower_range:self.row_higher_range, self.col_lower_range:self.col_higher_range]
            
        elif self.squre_number == 3:
            self.row_lower_range = 3
            self.row_higher_range = 6
            self.col_lower_range = 0
            self.col_higher_range = 3
                
            return self.board[self.row_lower_range:self.row_higher_range, self.col_lower_range:self.col_higher_range]
            
        elif self.squre_number == 4:
            self.row_lower_range = 3
            self.row_higher_range = 6
            self.col_lower_range = 3
            self.col_higher_range = 6
                
            return self.board[self.row_lower_range:self.row_higher_range, self.col_lower_range:self.col_higher_range]
            
        elif self.squre_number == 5:
            self.row_lower_range = 3
            self.row_higher_range = 6
            self.col_lower_range = 6
            self.col_higher_range = 9
                
            return self.board[self.row_lower_range:self.row_higher_range, self.col_lower_range:self.col_higher_range]
            
        elif self.squre_number == 6:
            self.row_lower_range = 6
            self.row_higher_range = 9
            self.col_lower_range = 0
            self.col_higher_range = 3
                
            return self.board[self.row_lower_range:self.row_higher_range, self.col_lower_range:self.col_higher_range]
            
        elif self.squre_number == 7:
            self.row_lower_range = 6
            self.row_higher_range = 9
            self.col_lower_range = 3
            self.col_higher_range = 6
                
            return self.board[self.row_lower_range:self.row_higher_range, self.col_lower_range:self.col_higher_range]
            
        elif self.squre_number == 8:
            self.row_lower_range = 6
            self.row_higher_range = 9
            self.col_lower_range = 6
            self.col_higher_range = 9
                
            return self.board[self.row_lower_range:self.row_higher_range, self.col_lower_range:self.col_higher_range]
            

game = Sudoku("417950030000000700060007000050009106800600000000003400900005000000430000200701580")
print(game.board)
print(game.get_row(0))
print(game.get_col(8))
print(game.get_sqr(1).flatten().tolist()) #➞ [9, 5, 0, 0, 0, 0, 0, 0, 7]
print(game.get_sqr(1, 8).flatten().tolist()) #➞ [0, 3, 0, 7, 0, 0, 0, 0, 0]
print(game.get_sqr(8, 3).flatten().tolist()) #➞ [0, 0, 5, 4, 3, 0, 7, 0, 1]

```

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
        self.lst = [val]
    
    def add_data(self, lst):
        """This function add data from last."""
        self.lst.extend(lst)
    
    def get_data(self):
        """this function return a list"""
        return self.lst
    
    def add_two_numbers(self, num1, num2):
        """this function return sum of two nodeList."""
        self.num1 = "".join([str(c) for c in num1.get_data()])
        self.num2 = "".join([str(c) for c in num2.get_data()])
        sum_listnode_obj = int(self.num1)+int(self.num2)
        a= list(str(sum_listnode_obj)[::-1])
        self.lst = [int(i) for i in a]
        return self.lst
    
try:
    lt1 = ListNode(9)
    lt1.add_data([9, 9, 9, 9, 9, 9])
    
    lt2 = ListNode(9)
    lt2.add_data([9,9,9])
    print(lt1.get_data())
    print(lt2.get_data())
    print(ListNode.add_two_numbers(lt1,lt1, lt2))
except Exception as e:
    print(e)
```

3. Write a class called CoffeeShop, which has three instance variables:

   1. name : a string (basically, of the shop)
   2. menu : a list of items (of dict type), with each item containing the item (name of the item), type (whether a food or a drink) and price.
   3. orders : an empty list

and seven methods:

  1. add_order: adds the name of the item to the end of the orders list if it exists on the menu, otherwise, return "This item is currently unavailable!"
  2. fulfill_order: if the orders list is not empty, return "The {item} is ready!". If the orders list is empty, return "All orders have been fulfilled!"
  3. list_orders: returns the item names of the orders taken, otherwise, an empty list.
  4. due_amount: returns the total amount due for the orders taken.
  5. cheapest_item: returns the name of the cheapest item on the menu.
  6. drinks_only: returns only the item names of type drink from the menu.
  7. food_only: returns only the item names of type food from the menu.

IMPORTANT: Orders are fulfilled in a FIFO (first-in, first-out) order.

```python
class CoffeeShop:
    def __init__(self):
        self.name = "coffeeshop"
        self.menu = [
            {"name":"tuna sandwich",
                "type":"food",
                "price":11
            },
            {"name":"ham and cheese sandwich",
                "type":"food",
                "price":12
            }, 
            {"name":"bacon and egg",
                "type":"food",
                "price":13
            },
            {"name":"steak",
                "type":"food",
                "price":14
            },
            {"name":"hamburger",
                "type":"food",
                "price":15
            },
            {"name":"cinnamon roll",
                "type":"food",
                "price":16
            },

            {"name":"orange juice",
                "type":"drink",
                "price":8
            },
            {"name":"lemonade",
                "type":"drink",
                "price":7
            }, 
            {"name":"cranberry juice",
                "type":"drink",
                "price":6
            },
            {"name":"pineapple juice",
                "type":"drink",
                "price":5
            },
            {"name":"lemon iced tea",
                "type":"drink",
                "price":4
            },
            {"name":"vanilla chai latte",
                "type":"drink",
                "price":3
            },
            {"name":"hot chocolate",
                "type":"drink",
                "price":2
            },
            {"name":"iced coffee",
                "type":"drink",
                "price":1
            }

        ]
        self.order = []
    
    def add_order(self, item_name):
        """adds the name of the item to the end of the orders list if it exists on the menu,
            otherwise, return "This item is currently unavailable!"""
        self.menu_items_name = [c['name'] for c in self.menu]
        if item_name in self.menu_items_name:
            self.order.append(item_name)
            return "Order added!"
        else:
            return "This item is currently unavailable!"
    
    def fulfill_order(self):
        """if the orders list is not empty, return "The {item} is ready!". If the orders list is empty,
            return "All orders have been fulfilled!"""
        if len(self.order)>0:
            a = self.order.pop(0)
            return f"The {a} is ready."
        else:
            return "All orders have been fulfilled!"
    
    def list_orders(self):
        """returns the item names of the orders taken, otherwise, an empty list."""
        return self.order
    
    def due_amount(self):
        """returns the total amount due for the orders taken."""

        amt_list = []
        for i in self.menu:
            for j in self.order:
                if i['name'] == j:
                    amt_list.append(i['price'])
        return sum(amt_list)
        
    def cheapest_item(self):
        """this function return of cheapest item which are present in menu list."""
        output_dict = {c['price']:c['name'] for c in self.menu}
        keys = min(list(output_dict.keys()))
        return output_dict[keys]
        
    def drinks_only(self):
        "This function return a list of item which are drinkable."
        return [c['name'] for c in self.menu if c['type']=="drink"]
    
    def food_only(self):
        "This function return a list of item which are food."
        return [c['name'] for c in self.menu if c['type']=="food"]
      
try:
    tcs = CoffeeShop()
    print(tcs.drinks_only())
    print(tcs.food_only())
    print(tcs.cheapest_item())
    print(tcs.add_order("hot cocoa"))
    print(tcs.add_order("cinnamon roll"))
    print(tcs.add_order("cinnamon roll"))
    print(tcs.list_orders())
    print(tcs.due_amount())
    print(tcs.fulfill_order())
    print(tcs.fulfill_order())
    print(tcs.fulfill_order())
    print(tcs.due_amount())
except Exception as e:
    print(e)
```
