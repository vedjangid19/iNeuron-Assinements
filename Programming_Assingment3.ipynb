### Que 1.	Write a Python Program to Check if a Number is Positive, Negative or Zero?

```python
def checkNumPosNeg(a):
    """this function return "Positive", "Negative" or "Zero" of Number"""
    
    if a == 0:
        return f"{a} is Zero."
    elif a > 0:
        return f"{a} is Positive."
    else:
        return f"{a} is Negative."


try:
    a = float(input("Enter Number : "))
    print(checkNumPosNeg(a))
except Exception as e:
    print("please enter valid number.")
```

### Que 3.	Write a Python Program to Check Leap Year?

Leap Year condition is : 

it is divisible by 4.

it is divisible by 4 and 100 than check year divisible by 400 if year is divisible by 400 it is leap year

```python
def checkLeapYear(year):
    """This function return year is leap year or not."""
    if((year % 4 ==0) and ((year % 100 != 0) or (year % 400 == 0))):
        return f"{year} is a leap year."
    else:
        return f"{year} not a leap year."
    
try:
    year = int(input("Enter a year : "))
    print(checkLeapYear(year))
    
except Exception as e:
    print("Please enter a valid year")
    
```

### Que 4.	Write a Python Program to Check Prime Number?

```python
from math import sqrt

def checkPrimeNumber(number):
    """this function return number is prime or not.
    In this program 1 not consider as prime number."""
    count = 0
    
    if number > 1:
        for i in range(2,number+1):
            if number == 2:
                count = count+1
                break
            elif (number % i == 0):
                count = count+1

        if count == 1:
            return f"{number} is prime number."
        else:
            return f"{number} is not prime number!!!"
    else:
        return f"{number} is not prime number."
    
    
    
try:
    number = int(input("Enter a Number : "))
    print(checkPrimeNumber(number))
    
except Exception as e:
    print("Please enter a valid Number")
    
```

### Que 5.	Write a Python Program to Print all Prime Numbers in an Interval of 1-10000?

```python
def checkPrimeNumber(number):
    """this function return number is prime or not.
    In this program 1 not consider as prime number."""
    count = 0
    
    if number > 1:
        for i in range(2,number+1):
            if number == 2:
                count = count+1
                break
            elif (number % i == 0):
                count = count+1

        if count == 1:
            return f"{number} is prime number."
        else:
            return f"{number} is not prime number!!!"
    else:
        return f"{number} is not prime number."
    
try:
    number_range = [x for x in range(1,10001)]
    number = list(map(checkPrimeNumber,number_range ))
    for i in number:
        print(i)
except Exception as e:
    print("Please enter a valid Number")
    
```
